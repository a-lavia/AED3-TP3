```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

## Test 1
Genero instancias con cantidad de gimnasios crecientes hasta la cantidad de paradas (p) - 
p oscila entre 1 y 7 - 
g va desde 1 a p - 
mochila fija = 6 - 
posiciones aleatorias - 
dificultad 50 - 
x = 30, y = 30

```python
test1H = pd.read_csv('CasosPerformanceTiempo/Caso1/resultadosH.csv')
test1BT = pd.read_csv('CasosPerformanceTiempo/Caso1/resultadosBT.csv')
#test1resBT = pd.read_csv('CasosPerformanceTiempo/Caso1/resCaminosBT.csv')
#test1resH = pd.read_csv('CasosPerformanceTiempo/Caso1/resCaminosH.csv')
```

```python
#Quiero comparar en cantidad de clocks y en performance

# Valores heuristica
paradas1 = 7
gymH1 = test1H['gimnasios']
lengym1 = len(gymH1)
distanciasH1 = test1H['distancia']
ciclosH1 = test1H['cantciclos']
# Valores backtracking
distanciasBT1 = test1BT['distancia']
ciclosBT1 = test1BT['cantciclos']

nodos1 = []
tiempobt1 = []
tiempoh1 = []
disth1 = []
distbt1 = []
diff1 = []
diff1b = []
gym1 = []

for i in range(21,28):
    nodos1.append(paradas1+gymH1[i])
    tiempobt1.append(ciclosBT1[i])
    tiempoh1.append(ciclosH1[i])
    disth1.append(distanciasH1[i])
    distbt1.append(distanciasBT1[i])
    diff1.append(ciclosBT1[i] - ciclosH1[i])
    diff1b.append(distanciasH1[i] - distanciasBT1[i])
    gym1.append(i-21+1)

# Calculo sobre diferencias
print "MAXIMOS TIEMPOS"
print max(diff1)
for i in range(0, len(diff1)):
    if diff1[i] == max(diff1):
        print "en el indice", i

print "MINIMOS/MAXIMOS DISTANCIA"
print "min", min(diff1b)
for i in range(0, len(diff1b)):
    if diff1b[i] == min(diff1b):
        print "min en el indice", i
print "en indice 2", diff1b[2]
print "max", max(diff1b)
for i in range(0, len(diff1b)):
    if diff1b[i] == max(diff1b):
        print "max en el indice", i
```

```python
# Comparacion de tiempos
df1 = pd.DataFrame()
df1['Cantidad de Gimnasios'] = gym1
df1['Tiempo BT'] = tiempobt1
df1['Tiempo H'] = tiempoh1
df1.plot(x='Cantidad de Gimnasios', logy=True)
#plt.axis((8,14,0,1.2e10))
plt.ylabel('Tiempos (en ciclos)')
#plt.show()
plt.savefig('ej2test1hvsbttiemposa.pdf', format='pdf', bbox_inches = 'tight', dpi = 100)
```

```python
# Comparaci贸n de resultados
plt.clf()
df1perf = pd.DataFrame()
df1perf['Cantidad de Gimnasios'] = gym1
df1perf['Distancia BT'] = distbt1
df1perf['Distancia H'] = disth1
#df1perf['Diferencias'] = diff1
df1perf.plot(x='Cantidad de Gimnasios')
#plt.axis((8,14,0,1.2e10))
#plt.show()
plt.savefig('ej2test1hvsbttiemposb.pdf', format='pdf', bbox_inches = 'tight', dpi = 100)
```
